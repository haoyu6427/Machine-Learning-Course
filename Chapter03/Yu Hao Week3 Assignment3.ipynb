## Step3 Using the Scikit-Learn Library, train the Logistic Regression model 

### Step3.1 All six cases of using two features at a time

```python
# -*- coding: utf-8 -*-
"""
Created on Wed Sep 18 17:03:12 2019

@author: Sumail
"""

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt
import numpy as np
import plot_decision_regions

iris = datasets.load_iris()
dict1 = [[0,1], [0,2], [0,3], [1,2], [1,3], [2,3]]
dict11 = []
dictc = []
dict12 = []
for i in range(len(dict1)):
    X = iris.data[:, dict1[i]]
    y = iris.target
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=1, stratify=y)
    
    sc = StandardScaler()
    sc.fit(X_train)
    X_train_std = sc.transform(X_train)
    X_test_std = sc.transform(X_test)
    
    X_combined_std = np.vstack((X_train_std, X_test_std))
    y_combined = np.hstack((y_train, y_test))
    
    
    lr = LogisticRegression(C=100.0, random_state=1, solver = 'liblinear', multi_class = 'auto')
    lr.fit(X_train_std, y_train)
    
    plot_decision_regions.plot_decision_regions(X_combined_std, y_combined,
                          classifier=lr, test_idx=range(105, 150))
    plt.xlabel('petal length [standardized]')
    plt.ylabel('petal width [standardized]')
    plt.legend(loc='upper left')
    plt.tight_layout()
    one_hot = OneHotEncoder(categories='auto',sparse=False)
    y_true = one_hot.fit_transform(y_test.reshape(-1,1))
    y_pred = one_hot.fit_transform(lr.predict(X_test_std).reshape(-1,1))
    logloss = log_loss(y_true, y_pred)
    #plt.savefig('images/03_06.png', dpi=300)
    print('2 features are [%d,%d]. iteration is %d. test set accuracy is %f. logloss is %f'%(dict1[i][0], dict1[i][1], lr.n_iter_, lr.score(X_test_std, y_test),logloss))
    dict11.append([dict1[i][0], dict1[i][1], lr.n_iter_, lr.score(X_test_std, y_test), logloss])
    dictc.append('[' +str(dict1[i][0])+','+str(dict1[i][1])+']')
    dict12.append(lr.score(X_test_std, y_test))
    plt.show()
for dict in dict11:
    print('2 features are [%d,%d]. iteration is %d. test set accuracy is %f. logloss is %f'%(dict[0], dict[1], dict[2], dict[3],dict[4]))
plt.figure(figsize= (10, 6))
plt.bar(range(len(dictc)), dict12, tick_label=dictc)
plt.title("2 Features Average Accuracy")
plt.xlabel("Features")
plt.ylabel("Accuracy")
plt.show()
```

## Step 6 Foundings

### Step 6.1 L1 and  L2 regularization: As for IRIS dataset, it takes 20-37 iterations for the model to converge using L1 regularization and takes 8-9 iterations usging L2 regularization. The accuracy of L1 and  L2 regularization is basically the same. Thus, Choosing the suitable regularization can increase the speed of training model.

### Step 6.2  The regularization parameter C: The regularization parameter C plays an important role in training process. As the parameter C increases, the number of iterations increases and the accuracy increases. When the parameter C is increased to a certain number, the number of iterations will increase but the accuracy will no longer increase.Thus, if the parameter C is too small, the accuracy will be low. If the parameter C is too big, we can get a high accuracy but the iterations will be too large so that the training process cost a lot of time.

### Step 6.3 Does using more dimensions help when trying to classify the data in this dataset? No. According to the result, using more dimensions doesn't increase the accuracy so that it fails to help when trying to classify the data in this dataset
